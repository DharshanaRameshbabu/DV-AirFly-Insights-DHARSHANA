```python
df['dep_time'] = pd.to_numeric(df['dep_time'], errors='coerce')
df['arr_time'] = pd.to_numeric(df['arr_time'], errors='coerce')
df['dep_hour'] = (df['dep_time'] // 100).astype('Int64')
df['dep_minute'] = (df['dep_time'] % 100).astype('Int64')
df['arr_hour'] = (df['arr_time'] // 100).astype('Int64')
df['arr_minute'] = (df['arr_time'] % 100).astype('Int64')
print(df[['dep_time', 'arr_time', 'dep_hour', 'dep_minute', 'arr_hour', 'arr_minute']].head(10))
```

```python
df['flight_duration'] = (
    (df['arr_hour'] * 60 + df['arr_minute']) -
    (df['dep_hour'] * 60 + df['dep_minute'])
)
print(df[['dep_time', 'arr_time', 'dep_hour', 'dep_minute', 'arr_hour', 'arr_minute', 'flight_duration', 'air_time']].head(10))
```

```python

df['flight_date'] = pd.to_datetime(df[['year', 'month', 'day']])
df['month'] = df['flight_date'].dt.month
df['day_of_week'] = df['flight_date'].dt.dayofweek
df['is_weekend'] = df['day_of_week'].isin([5, 6]).astype(int)  
df['hour'] = df['dep_hour'] 
df['route'] = df['origin'] + '-' + df['dest']
df[['flight_date', 'month', 'day_of_week', 'is_weekend', 'hour', 'route']].head()
```

```python
df.to_csv('flights_cleaned.csv', index=False)
df.to_parquet('flights_cleaned.parquet', index=False)
```

```python
summary_text = """
- Dataset size: {} rows, {} columns.
- Missing values handled for dep_time, arr_time, dep_delay, arr_delay, air_time, and tailnum.
- Created new datetime column 'flight_date' from year, month, day.
- Engineered features:
  - month: Numeric month of flight.
  - day_of_week: Weekday number (0=Monday, 6=Sunday).
  - is_weekend: Flag for Saturday/Sunday flights (1 = weekend).
  - hour: Departure hour.
  - route: A string combining origin and destination airport codes.
- Calculated flight_duration from departure and arrival times.
""".format(df.shape[0], df.shape[1])
with open('preprocessing_summary.md', 'w') as f:
    f.write(summary_text)
print("Preprocessing summary saved.")
```

```python
feature_dict = ""

| Feature         | Description                                  | Type       |
|-----------------|----------------------------------------------|------------|
| flight_date     | Date of the flight                            | datetime   |
| month           | Month of flight (1 to 12)                     | integer    |
| day_of_week     | Day of the week (Monday=0 to Sunday=6)       | integer    |
| is_weekend      | Weekend flag (1 if Saturday or Sunday)       | integer    |
| hour            | Hour of departure (0-23)                      | integer    |
| route           | Origin and destination airport codes joined  | string     |
| dep_time        | Scheduled departure time (HHMM)               | integer    |
| arr_time        | Scheduled arrival time (HHMM)                 | integer    |
| dep_delay       | Departure delay in minutes                     | float      |
| arr_delay       | Arrival delay in minutes                       | float      |
| flight_duration | Calculated flight duration in minutes         | integer    |
"""
with open('feature_dictionary.md', 'w') as f:
    f.write(feature_dict)
print("Feature dictionary saved.")
```
