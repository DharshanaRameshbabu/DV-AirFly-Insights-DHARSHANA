```python
def minutes_to_hours_minutes(duration):
    hours, mins = divmod(duration, 60)
    return f"{int(hours)}:{int(mins):02d}"

df['flight_duration_formatted'] = df['flight_duration'].apply(minutes_to_hours_minutes)
```

```python
df['dep_time'] = pd.to_numeric(df['dep_time'], errors='coerce')
df['arr_time'] = pd.to_numeric(df['arr_time'], errors='coerce')
df['dep_hour'] = (df['dep_time'] // 100).astype('Int64')
df['dep_minute'] = (df['dep_time'] % 100).astype('Int64')
df['arr_hour'] = (df['arr_time'] // 100).astype('Int64')
df['arr_minute'] = (df['arr_time'] % 100).astype('Int64')
print(df[['dep_time', 'arr_time', 'dep_hour', 'dep_minute', 'arr_hour', 'arr_minute']].head(10))
```

```python
df['flight_duration'] = (
    (df['arr_hour'] * 60 + df['arr_minute']) -
    (df['dep_hour'] * 60 + df['dep_minute'])
)
print(df[['dep_time', 'arr_time', 'dep_hour', 'dep_minute', 'arr_hour', 'arr_minute', 'flight_duration', 'air_time']].head(10))
```

```python

df['flight_date'] = pd.to_datetime(df[['year', 'month', 'day']])
df['month'] = df['flight_date'].dt.month
df['day_of_week'] = df['flight_date'].dt.dayofweek
df['is_weekend'] = df['day_of_week'].isin([5, 6]).astype(int)  
df['hour'] = df['dep_hour'] 
df['route'] = df['origin'] + '-' + df['dest']
df[['flight_date', 'month', 'day_of_week', 'is_weekend', 'hour', 'route']].head()
```

```python
df.to_csv('flights_cleaned.csv', index=False)
df.to_parquet('flights_cleaned.parquet', index=False)
```

```python
summary_text = """
- Dataset size: {} rows, {} columns.
- Missing values handled for dep_time, arr_time, dep_delay, arr_delay, air_time, and tailnum.
- Created new datetime column 'flight_date' from year, month, day.
- Engineered features:
  - month: Numeric month of flight.
  - day_of_week: Weekday number (0=Monday, 6=Sunday).
  - is_weekend: Flag for Saturday/Sunday flights (1 = weekend).
  - hour: Departure hour.
  - route: A string combining origin and destination airport codes.
- Calculated flight_duration from departure and arrival times.
""".format(df.shape[0], df.shape[1])
with open('preprocessing_summary.md', 'w') as f:
    f.write(summary_text)
print("Preprocessing summary saved.")
```

```python
feature_dict = """
| Feature         | Description                                  | Type       |
|-----------------|----------------------------------------------|------------|
| flight_date     | Date of the flight                            | datetime   |
| month           | Month of flight (1 to 12)                     | integer    |
| day_of_week     | Day of the week (Monday=0 to Sunday=6)       | integer    |
| is_weekend      | Weekend flag (1 if Saturday or Sunday)       | integer    |
| hour            | Hour of departure (0-23)                      | integer    |
| route           | Origin and destination airport codes joined  | string     |
| dep_time        | Scheduled departure time (HHMM)               | integer    |
| arr_time        | Scheduled arrival time (HHMM)                 | integer    |
| dep_delay       | Departure delay in minutes                     | float      |
| arr_delay       | Arrival delay in minutes                       | float      |
| flight_duration | Calculated flight duration in minutes         | integer    |
"""

with open('feature_dictionary.md', 'w') as f:
    f.write(feature_dict)

print("Feature dictionary saved.")
```

```python
#milestone 2
import pandas as pd
import numpy as np
import matplotlib.pyplot as plt
import seaborn as sns
sns.set_style("whitegrid")
```

```python
# Busiest Months
monthly_flights = df['month'].value_counts().sort_index()
plt.figure(figsize=(10, 6))
sns.lineplot(x=monthly_flights.index, y=monthly_flights.values, marker='o', color='skyblue')
plt.title('Flight Distribution by Month')
plt.xlabel('Month')
plt.ylabel('Number of Flights')
plt.xticks(np.arange(1, 13))
plt.grid(True)
plt.show()
```

```python
#Top Airlines
top_airlines = df['name'].value_counts().head(10).sort_values()
plt.figure(figsize=(10, 6))
sns.scatterplot(x=top_airlines.values, y=top_airlines.index, s=150, color='blue', zorder=2)
plt.plot(top_airlines.values, top_airlines.index, linestyle='--', color='gray', linewidth=0.5)
plt.title('Top 10 Airlines by Flight Count')
plt.xlabel('Number of Flights')
plt.ylabel('Airline Name')
plt.grid(axis='x', zorder=1)
plt.tight_layout()
plt.show()
```

```python
#Top Routes
df['route'] = df['origin'] + ' -> ' + df['dest']
top_routes = df['route'].value_counts().head(10)
plt.figure(figsize=(12, 6))
sns.barplot(x=top_routes.values, y=top_routes.index, palette='plasma', hue=top_routes.index, legend=False)
plt.title('Top 10 Busiest Routes')
plt.xlabel('Number of Flights')
plt.ylabel('Route (Origin -> Destination)')
plt.tight_layout()
plt.show()
```

```python
#Flight Distribution by Time of Day 
plt.figure(figsize=(10, 6))
sns.histplot(df['hour'], bins=24, kde=False, color='green')
plt.title('Flight Distribution by Hour of Day')
plt.xlabel('Hour of Departure (0-23)')
plt.ylabel('Number of Flights')
plt.xticks(np.arange(0, 24, 2))
plt.show()
```

```python
#Arrival Delay Distribution by Month
# Calculate the median arrival delay for each month
monthly_median_delay = df.groupby('month')['arr_delay'].median()

plt.figure(figsize=(12, 6))
sns.lineplot(
    x=monthly_median_delay.index,
    y=monthly_median_delay.values,
    marker='o',
    color='darkred'
)

plt.title('Median Arrival Delay Trend by Month')
plt.xlabel('Month')
plt.ylabel('Median Arrival Delay (Minutes)')
plt.xticks(np.arange(1, 13))
plt.grid(True)
plt.show()
```

```python
#Average Delay by Airline
avg_delay_carrier = df.groupby('name')['arr_delay'].mean().sort_values(ascending=False).head(10)
plt.figure(figsize=(12, 6))
sns.barplot(x=avg_delay_carrier.values, y=avg_delay_carrier.index, palette='plasma', hue=avg_delay_carrier.index, legend=False)
plt.title('Average Arrival Delay by Top 10 Airlines')
plt.xlabel('Average Arrival Delay (Minutes)')
plt.ylabel('Airline Name')
plt.tight_layout()
plt.show()
```

```python
#Relationship Between Distance and Air Time
plt.figure(figsize=(10, 6))
# Plot the distribution of flight distances
sns.histplot(df['distance'], bins=50, kde=True, color='purple')

plt.title('Distribution of Flight Distances')
plt.xlabel('Distance (Miles)')
plt.ylabel('Frequency')
plt.xlim(0, 5000) # Set reasonable limits for distance
plt.show()
```

```python
# Delay by Time of Day 
plt.figure(figsize=(14, 8))

# Use Strip Plot with 'jitter' to display all data points for each hour
sns.stripplot(x='hour', 
              y='arr_delay', 
              data=df.loc[(df['arr_delay'] > -30) & (df['arr_delay'] < 120)], # Filter for clarity
              jitter=0.3, # Adds a slight horizontal spread so points don't stack perfectly
              size=3,     # Small size for a large dataset
              color='darkred', 
              alpha=0.5)

plt.title('Individual Flight Delay Distribution by Hour of Day (Strip Plot)')
plt.xlabel('Hour of Departure')
plt.ylabel('Arrival Delay (Minutes)')
plt.grid(True)
plt.show()
```
